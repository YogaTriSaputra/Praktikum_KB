# Import Library


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Membaca File CSV Kemudian Disimipan Kedalam Sebuah Variabel "pes"

```python
pes = pd.read_csv("pes2021-all-players.csv")
```

# Grafik Pie

Grafik Pie cocok digunakan pada data categorical. Grafik ini biasa digunakan untuk melihat distribusi.

Grafik Pie dibawah ini menunjukkan perbandingan persentase jumlah player/pemain yang memakai kaki kanan dan kaki kiri. Terlihat bahwa player/pemain dengan memakai kaki kanan lebih mendominasi 55,2% di bandingkan dengan player/pemain dengan memakai kaki kiri.

```python
plt.figure(figsize=(10,10))
plt.pie(pes['foot'].value_counts(), labels=['Right Foot', 'Left Foot'], startangle=90, autopct='%.1f%%', shadow = True, explode=(0.1,0))
plt.title('Grafik Pie', fontsize = 20, fontweight = 'bold')
plt.legend()
plt.show()
```

Grafik Pie di bawah menunjukkan bahwa player/pemain yang mempunyai rating bintang 1 lebih banyak yaitu 59,6% dan palyer/pemain yang mempunyai rating bintang 5 paling sedikit yaitu 0,8%.

```python
plt.figure(figsize = (20,20))
plt.pie(pes['rating_stars'].value_counts(), labels = pes['rating_stars'].value_counts().keys().to_list(), startangle = 90, autopct = '%.1f%%')
plt.title('Grafik Pie',fontsize = 20, fontweight='bold')
plt.legend()
plt.show()
```

# Grafik Scatter Plot

Scatter plot menunjukan hubungan antara dua variabel. Scatter plot seringdigunakan sebagai analisis tindak lanjut untuk menentukan apakah penyebab yang ada benar-benar memberikan dampak kepada karakteristik kualitas.

Grafik Scatter Plot di bawah ini menunjukan player/pemain terberat adalah GK dengan berat kurang lebih 110 ke atas. Sedangkan pemain terkurus adalah LB dengan berat kurang lebih 40 ke bawah.

```python
plt.scatter(pes['registered_position'], pes['weight'], label = 'Berat',color  = 'orange')
plt.title('Grafik Scatter Plot',fontsize = 20 , fontweight = 'bold')
plt.xticks(rotation = 90)
plt.xlabel('Position')
plt.ylabel('Weight')
plt.legend()
plt.show()
```

# Grafik Batang

Grafik batang sangat cocok untuk memvisualisasi data kategorikal. Grafik ini biasa digunakan untuk membandingkan kategori.

Grafik batang di bawah ini menunjukan bahwa region dengan jumlah player/pemain terbanyak adalah Europe dengan jumlah kurang lebih 7000 ke atas. Sedangkan region dengan jumlah player/pemain tersedikit adalah North & Central America.

```python
plt.figure(figsize = (10,10))
pes['region'].value_counts().plot.bar()
plt.title('Grafik Batang',fontsize = 20 , fontweight = 'bold')
plt.xlabel('region')
plt.ylabel('Jumlah')
plt.show()
```
