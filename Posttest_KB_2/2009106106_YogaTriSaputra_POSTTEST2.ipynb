```python
import pandas as pd
```

Menyimpan dataset dalam sebuah variabel "pes"

```python
pes = pd.read_csv("pes2021-all-players.csv")
```

```python
pes
```

Informasi jumlah record, nama attribute, jumlah attribute, tipe objek, jumlah attribute, tipe angka, jumlah attribute yang memiliki nilai null ke dalam dictionary. Sekaligus menampilkan isi dari dictionary tersebut.

```python
jumlahRecord = len(pes)
namaAttribute = list(pes.columns)
jumlahAttributeObject = len(pes.select_dtypes(include = ["object"]).columns)
jumlahAttributeAngka = len(pes.select_dtypes(include = ["number"]).columns)
jumlahAttributeNull = len(pes.columns[pes.isna().sum() > 0])

dictionaryPes = {
    "Jumlah Record             :": jumlahRecord,
    "Nama Attribute            :": namaAttribute,
    "Jumlah Attribute Object   :": jumlahAttributeObject,
    "Jumlah Attribute Angka    :": jumlahAttributeAngka,
    "Jumlah Attribute Null     :": jumlahAttributeNull,
}

pd.Series(dictionaryPes)
```
